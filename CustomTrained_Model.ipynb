```python
import os

# Path to your coke dataset directory
dataset_dir = '/content/coke-dataset/valid/labels'

# List all label files in the dataset directory
label_files = [file for file in os.listdir(dataset_dir) if file.endswith('.txt')]

# Iterate through each label file and replace label 0 with label 1
for label_file in label_files:
    file_path = os.path.join(dataset_dir, label_file)
    with open(file_path, 'r+') as f:
        lines = f.readlines()
        f.seek(0)
        for line in lines:
            parts = line.strip().split()
            if len(parts) > 0:
                label = int(parts[0])
                if label == 0:
                    parts[0] = '1'
                f.write(' '.join(parts) + '\n')
        f.truncate()
    print(f"Updated labels in {label_file}")

print("Label update complete.")
```

```python
import os
import shutil

# Define paths to your datasets
source_dataset_dir = '/content/coke-dataset'   # Path to coke-dataset
destination_dataset_dir = '/content/pepsi-dataset'   # Path to pepsi-dataset

# Ensure the destination directory exists, create it if it doesn't
os.makedirs(destination_dataset_dir, exist_ok=True)

# Function to copy all files from source to destination
def copy_dataset(source_dir, destination_dir):
    for root, dirs, files in os.walk(source_dir):
        for file in files:
            source_file = os.path.join(root, file)
            dest_file = os.path.join(destination_dir, os.path.relpath(source_file, source_dir))
            dest_dir = os.path.dirname(dest_file)
            os.makedirs(dest_dir, exist_ok=True)
            shutil.copyfile(source_file, dest_file)

# Copy the dataset
copy_dataset(source_dataset_dir, destination_dataset_dir)

print("Dataset copied successfully from coke-dataset to pepsi-dataset.")
```
