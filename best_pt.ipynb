```python
!pip install ipywidgets

from IPython.display import HTML

video_path = '/content/Coke_Pepsi_Video.mp4'

# Generate HTML code to display the video
HTML(f'<video controls src="{video_path}" width="100%" height="auto">')
```

```python
import cv2
import json
from ultralytics import YOLO

# Load the custom-trained YOLOv8 model
model_path = "/content/best.pt"
model = YOLO(model_path)

# Input and output video paths
input_video_path = '/content/Coke_Pepsi_Video.mp4'
output_video_path = 'output.mp4'

# Open the input video
cap = cv2.VideoCapture(input_video_path)
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_video_path, fourcc, 30.0, (int(cap.get(3)), int(cap.get(4))))

# Dictionary to hold detection timestamps
detection_timestamps = {"Pepsi": [], "Coke": []}

# Get the frames per second of the video
fps = cap.get(cv2.CAP_PROP_FPS)

while cap.isOpened():
    frame_number = int(cap.get(cv2.CAP_PROP_POS_FRAMES))
    ret, frame = cap.read()
    if not ret:
        break

    # Perform detection
    results = model(frame)

    # Calculate the timestamp
    timestamp = frame_number / fps

    # Draw bounding boxes and labels on the frame
    for result in results:
        for box in result.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            confidence = box.conf[0]
            class_id = int(box.cls[0])
            label = model.names[class_id]

            # Log the detection timestamp if it's Pepsi or Coke
            if label.lower() == "pepsi":
                detection_timestamps["Pepsi"].append(timestamp)
            elif label.lower() == "coca-cola":
                detection_timestamps["Coke"].append(timestamp)

            # Draw the bounding box and label
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            if label == "pepsi":
                cv2.putText(frame, f'{"Coke"} {confidence:.2f}', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
            else:
                cv2.putText(frame, f'{"Pepsi"} {confidence:.2f}', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
            # cv2.putText(frame, f'{label} {confidence:.2f}', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    # Write the frame with detections to the output video
    out.write(frame)

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()

# Save the detection timestamps to a JSON file
with open('result.json', 'w') as f:
    json.dump(detection_timestamps, f, indent=4)

print("Detection complete. Output video saved at:", output_video_path)
print("Detection timestamps saved in result.json")
```
